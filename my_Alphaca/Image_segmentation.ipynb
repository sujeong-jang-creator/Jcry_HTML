## Train segmentation

```python
# 메모리 문제로 1000개 우선저장
x_train = []
y_train = []
x_train_nor = []
y_train_num = []

for idx, c in enumerate(categories):
    print(idx)
    label = [0 for i in range(nb_class)]
    label[idx] = 1

    image_dir = train_data_dir+"/Image_"+c
    json_dir = train_data_dir+"/Label_"+c
    files = glob.glob(image_dir+"/*.jpg")
    json_files = glob.glob(json_dir+"/*.json")
    
    print(c)
    # 이미지 파일과 json 파일이 같은 순서로 나오는 것 확인하는 코드
    # for i in range(len(files)):
    #    print(json_files[i])
    #    print(files[i])
    cnt = 0;
    for i, f in enumerate(files):
        if cnt >= 1000:
            break
        cnt += 1
        json_data = dict()
        with open(json_files[i]) as jf:
            json_data = json.load(jf)
            
        contour = json_data['label_info']['shapes'][0]['points']
        
        min_x = 300000
        min_y = 300000
        max_x = -1
        max_y = -1
        
        for i in range(len(contour)):
            contour[i][0] = int(contour[i][0])
            contour[i][1] = int(contour[i][1])
            if contour[i][0] > max_x:
                max_x = contour[i][0]
            if contour[i][0] < min_x:
                min_x = contour[i][0]
            if contour[i][1] > max_y:
                max_y = contour[i][1]
            if contour[i][1] < min_y:
                min_y = contour[i][1]
         
        
        img = Image.open(f)
        img = img.convert("RGB")
        img = np.asarray(img)
        
        mask = np.zeros_like(img)

        polygon = np.array(contour)
        
        cv2.fillPoly(mask,[polygon],[255, 255, 255])
        masked = cv2.bitwise_and(img, mask)
        masked_crop = masked[min_y : max_y+1, min_x :max_x+1]
#         print(type(masked_crop))
#         print(masked_crop.shape) # (645, 454, 3)

        masked_crop = cv2.resize(masked_crop,(image_w,image_h))
#         print(type(masked_crop))
#         print(masked_crop.shape)
        masked_crop = masked_crop.astype('float32')/255
        x_train.append(masked_crop)
        y_train.append(c)
        y_train_num.append(idx)
        
        
        
# x_train = np.array(x_train)
# y_train = np.array(y_train)

# x_train_norm =  x_train.astype('float32')/255
```

```python
# array 형식으로 변환
x_train = np.array(x_train)
y_train = np.array(y_train)
y_train_num = np.array(y_train_num)
```
