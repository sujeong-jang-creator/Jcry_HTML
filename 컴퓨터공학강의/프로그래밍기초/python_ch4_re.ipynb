```python
def calculate_change(payment, cost):
    money = payment - cost
    unit = [50000,10000,5000,1000]
    for i in range(len(unit)):
        print('{}원 지페 : {}장'.format(unit[i],money//unit[i]))
        money = money % unit[i]
calculate_change(100000, 33000) # 함수 호출
print('=================')
calculate_change(500000, 378000)
```

```python
def abs(num):
    # return만나면 함수가 종료됨
    if num > 0:
        result =  num
    else:
        result =  num * -1
    return result
print(abs(-7))
print(abs(12))
```

```python
# 두수를 받아서 사칙연산결과를 출력
def cal(num1, num2):
    print("{} + {} = {}".format(num1,num2,num1+num2))
    print("{} - {} = {}".format(num1, num2, num1 - num2))
    print("{} * {} = {}".format(num1, num2, num1 * num2))
    print("{} / {} = {}".format(num1,num2,num1/num2))

cal(12, 3)
```

```python
x = 20; y = 10
k = [1,2,3]
# call함수에서 x,y를 받아서 x+=y를 실행한후에
# 함수 안에서 x, y출력, 함수 밖에서 x,y출력하고 비교
# k에 4를 추가하고 함수 안과 밖에서 출력
def call(x, y, k):
    x += y
    k.append(4)
    print('x =',x,', y =',y)
    print(k)
call(x,y,k)
print('x =',x,', y =',y)
print(k)
```

```python
def multiply_by_two():
    global x
    x = x * 2
def multiply_by_three():
    y = 2  # local
    y = y * 3
x = 2
multiply_by_two()
# x는 global이므로 변경된 4가 출력
print(x)
y = 2
multiply_by_three()
# y는 지역함수에서 변경이 내용이 반영되지 않으므로 2
print(y)
```

```python
def is_evenly_divisible(number):
    # 짝수면 True, 홀수면 False
    return number % 2 == 0

print(is_evenly_divisible(3))
print(is_evenly_divisible(7))
print(is_evenly_divisible(8))
```

```python
def is_odd(num):
    if num % 2 == 0:
        result = "짝수"
    else:
        result = "홀수"
    return result
print(is_odd(3))
print(is_odd(7))
print(is_odd(8))
```

```python
# 입력으로 들어오는 모든 수의 평균 값을 계산해 주는 함수를 작성
# 단 입력으로 들어오는 수의 개수는 정해져 있지 않다.
# ※ 평균 값을 구할 때 len 함수를 사용해 보자
def average(*val):
    sum = 0
    for i in val:
        sum += i
    print("합계 : {0:d}, 평균 : {1:.2f}".
          format(sum, sum/len(val)))
average(45, 67, 88)
average(78, 99, 88, 66)
```
