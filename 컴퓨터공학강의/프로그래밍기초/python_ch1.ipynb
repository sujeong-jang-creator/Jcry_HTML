if문 연습

```python
x = -100
# 콜론(:)는 조건이 끝났다는 의미
if x > 0:
    # 들여쓰기가 상위 문장의 조건을 만족할 때 실행한다는 의미
    print("0보다 큽니다")
    # 세미콜론(;)은 두개이상 문장을 한 줄로 쓸때 사용
    print("x = ",x); print("대박")
else :
    print("0보다 작거나 같다")
    print("헐 ~")
```

대입과 if문 

```python
a1 = 7
a2 = 10
if a1 > a2 :
    print("a1이 a2보다 크다")
else :
    print("a1이 a2보다 작거나 같다")
```

기본연산

```python

# /나누기 결과는 실수, //몫
k = a = 7  # =는 오른 쪽 부터 실행
# print 안에서 ,는 한쪽 떨어짐
print('k =',k,", a =",a)
a1, b1 = 7, 8 # a1에는 7이 b1에는 8이 각각 대입
print('a1 =',a1,', b1 =',b1)
a1, b1 = b1, a1 # 값교환
print('a1 =',a1,', b1 =',b1)
```

대입과 출력

```python
# 15를 a1, a2, a3에 대입하고 출력
a1=a2=a3=15
print('a1 =',a1,', a2 =',a2,', a3 =',a3)
# b1, b2에 23과 45를 대입하고 출력
b1, b2 = 23, 45
print('b1 =',b1,', b2 =',b2)
# b1과 b2의 값을 서로 교환하고 출력
b1, b2 = b2, b1
print('b1 =',b1,', b2 =',b2)
# 허걱을 10회 출력
print('허걱'*10)
# 15나누기 3이 몫을 출력
print(15//3) # 몫은 정수
print(15/3)  # 나누기 결과는 실수
# 3의 3승을 출력
print(3**3)
```
